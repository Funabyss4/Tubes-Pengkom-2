```python
import matplotlib.pyplot as plt
import matplotlib.dates as dts
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
```

Mengambil data

```python
dfChihuahua = pd.read_csv("C:/Users/surya/Downloads/Ecoregions_a1b_Temp_ECHAM5-MPI.csv", sep=',')
dfChihuahua['TIMESTEP'] = pd.to_datetime(dfChihuahua['TIMESTEP'], format='%m/%d/%Y')
```

Plotting

```python
plt.figure(figsize=(10, 8))
plt.title("Grafik garis suhu rata-rata Chihuhuan Deserts")

# Chihuahua
# Suhu rata-rata
plt.plot(dfChihuahua.dropna()['TIMESTEP'], dfChihuahua.dropna()['Chihuahuan Deserts'])


plt.xlabel('TIMESTEP')
plt.ylabel('Suhu (\u00b0C)')
plt.legend(['Suhu rata-rata Chihuahuan Deserts'])
plt.show()
```
