```python
import netCDF4 as nc
from matplotlib.colors import LightSource
import matplotlib.pyplot as plt
import numpy as np
```

Mengambil data

```python
data = r'C:\Users\surya\Downloads\global-analysis-forecast-phy-001-024_1668127733073.nc'
fh = nc.Dataset(data)

lon = np.array(fh.variables['longitude'][:], dtype=np.float32)
lat = np.array(fh.variables['latitude'][:], dtype=np.float32)
time = np.array(fh.variables['time'][:], dtype=np.float32)
depth = np.array(fh.variables['depth'][:], dtype=np.float32)
mlotst = np.array(fh.variables['mlotst'][0,:], dtype=np.int16)
```

Plotting

```python
plt.figure()

ls = LightSource(azdeg=10000,altdeg = 1000)
i = 0

plt.title("Densitas laut")

for x in range(1000):
    i += 1
    plt.imshow(ls.shade(mlotst, cmap=plt.cm.gist_earth, vert_exag=10, blend_mode='hsv'), origin='lower')
    #plt.savefig(r"C:\Users\surya\OneDrive\Documents\map1"+'\\'+str(i+1) + '.jpg')

```
